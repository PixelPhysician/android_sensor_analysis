```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# For this experiment three different available light sensors were used
# These included TCS3701 Light, TCS3701 Light CCT	and camera light sensor

# These experiments were simulated at the FHNW by putting the phone towards a window (bright light) and shielding it with hand
```

```python
# Load the JSON file
with open('/content/DifferentLightSensors.json', 'r') as f:
    raw_data = json.load(f)

# Flatten the nested structure
# Each entry has a top-level with 'data' being the key of interest
records = []
for entry in raw_data:
    base = {k: v for k, v in entry.items() if k != "data"}
    data_fields = entry["data"]
    combined = {**base, **data_fields}
    records.append(combined)

# Convert to DataFrame
df = pd.DataFrame(records)
df['value'] = df['data'].apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None)


# Show table info and first rows
print(" DataFrame Shape:", df.shape)
print("\n Columns:", df.columns.tolist())
print("\n Head:")
display(df.head(50))
```

```python
# Show unique sensorId values
unique_sensor_ids = df['sensorId'].unique()
print("Unique sensorId values:")
for sid in unique_sensor_ids:
    print("-", sid)
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values('timestamp')

# Filter sensors
df_light = df[df['sensorId'] == 'TCS3701 Light']
df_cct = df[df['sensorId'] == 'TCS3701 Light CCT']

# Plot
plt.figure(figsize=(14, 5))
plt.plot(df_light['timestamp'], df_light['value'], color='green', label='TCS3701 Light')
plt.plot(df_cct['timestamp'], df_cct['value'], color='orange', label='TCS3701 Light CCT')
plt.xlabel("Time")
plt.ylabel("Sensor Reading (first value)")
plt.title("Light Sensor vs. CCT over Time")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Initial High Plateau (~600–700 lux) = Phone was facing daylight at the window.

Sudden Drop = shielded the sensor with your hand, blocking the light.

Flat Low Section = phone remained covered for several seconds, very low light.

Abrupt Return to High Values = removed hand, letting daylight hit the sensor again.

Another Drop = re-shielded the phone, repeating the hand coverage.
